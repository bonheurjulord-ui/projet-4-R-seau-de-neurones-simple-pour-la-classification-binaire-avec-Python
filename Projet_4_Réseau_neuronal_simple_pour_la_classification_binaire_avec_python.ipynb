**Étape 1 : Comprendre le problème (XOR)**

Le XOR est un classique :

- Entrées : 2 variables
- Sortie : 1 si elles sont différentes, sinon 0

```python
X = [0,0] → 0
X = [0,1] → 1
X = [1,0] → 1
X = [1,1] → 0
```

```python
import numpy as np
```

Créer le dataset

```python
# Entrées
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# Sorties
y = np.array([
    [0],
    [1],
    [1],
    [0]
])
```

**Initialiser le réseau**

Architecture demandée :

- 2 entrées
- 4 neurones cachés
- 1 sortie

```python
np.random.seed(42)

# Poids
W1 = np.random.rand(2, 4)  # entrée → couche cachée
b1 = np.zeros((1, 4))

W2 = np.random.rand(4, 1)  # cachée → sortie
b2 = np.zeros((1, 1))
```

**Fonction d’activation**

On utilise la sigmoïde :

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)
```

Entraînement (forward + backward)

```python
learning_rate = 0.1
epochs = 10000

for epoch in range(epochs):

    ## Forward pass
    z1 = np.dot(X, W1) + b1
    a1 = sigmoid(z1)

    z2 = np.dot(a1, W2) + b2
    a2 = sigmoid(z2)

    ## Calcul de la perte (MSE)
    loss = np.mean((y - a2) ** 2)

    ## Backpropagation
    d_a2 = (a2 - y)
    d_z2 = d_a2 * sigmoid_derivative(a2)

    d_W2 = np.dot(a1.T, d_z2)
    d_b2 = np.sum(d_z2, axis=0, keepdims=True)

    d_a1 = np.dot(d_z2, W2.T)
    d_z1 = d_a1 * sigmoid_derivative(a1)

    d_W1 = np.dot(X.T, d_z1)
    d_b1 = np.sum(d_z1, axis=0, keepdims=True)

    ## Mise à jour des poids
    W1 -= learning_rate * d_W1
    b1 -= learning_rate * d_b1

    W2 -= learning_rate * d_W2
    b2 -= learning_rate * d_b2

    ## Affichage perte
    if epoch % 1000 == 0:
        print(f"Epoch {epoch}, Loss: {loss}")
```

Tester le modèle

```python
print("Prédictions finales :")
print(a2)
```
